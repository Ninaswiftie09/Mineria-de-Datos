```python
#librerias 
import os 
import pyreadr
import pandas as pd 
import numpy as np 
import os 
import matplotlib.pyplot as plt
import seaborn as sns 

```

```python
#dataset 
base_path = os.getcwd()
data_path = os.path.join(base_path, "data", "listings.RData")

result = pyreadr.read_r(data_path)
df = result["listings"]

print("Dimensiones originales:", df.shape)
```

El conjunto de datos contiene 171,148 observaciones y 80 variables. 
Se identifican variables númericas, categóricas y de texto.
La variable objetivo es price, la cual se encuentra como tipo string y presenta valores faltantes. 

```python
#eliminación de filas sin variables respuesta 

df= df.dropna(subset=["price"])
print ("Dimensiones eliminando NA en price", df.shape)
```

Se eliminarion las observaciones sin precio ya que la variable respuesta no puede imputarse sin introducir sesgo en el modelo. 

```python
# Limpiar símbolos
df["price"] = (
    df["price"]
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.strip()
)

# Convertir a numérico
df["price"] = pd.to_numeric(df["price"], errors="coerce")

# Ver cuántos NA quedaron
print("NA después de convertir:", df["price"].isna().sum())
```

```python
df = df.dropna(subset=["price"])

print("Filas finales:", df.shape[0])
```

```python
df["price"].describe()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["price"], bins=50)
plt.title("Distribucion del precio")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df["price"])
plt.title("Boxplot del precio")
plt.show()
```

La variable precio presenta una distribución altamente asimétrica positiva, con una gran concentración de valores bajos y una cola larga hacia la derecha.
Esto indica la presencia de outliers y valores extremos, lo cual puede afectar el ajuste de modelos de regresión lineal.

```python
#transformacion logaritmica 
df["log_price"] = np.log1p(df["price"])

plt.figure(figsize=(8,5))
sns.histplot(df["log_price"], bins=50)
plt.title("Distribución del Log(Price)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df["log_price"])
plt.title("Boxplot del Log(Price)")
plt.show()
```
